```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("sinx.csv")
x = df["Input"]
y = df["Output"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential()

classifier.add(Dense(output_dim = 5, input_dim = 1, activation='relu', init = 'uniform'))

classifier.add(Dense(output_dim = 5, activation='linear', init='uniform'))

classifier.add(Dense(output_dim=1, activation='linear', init='uniform'))

classifier.compile(optimizer='SGD', loss='mean_squared_error', metrics=['mean_squared_error'])

classifier.fit(x, y, batch_size=40, nb_epoch=1000)
```

```python
# Evaluate the model
scores = classifier.evaluate(x, y)
print('%s: %.2f%%' % (classifier.metrics_names[1], scores[1] * 100))
```
